KNN smoothing on counts, adapting code from scvelo

```python
from scvelo.preprocessing.neighbors import get_connectivities
from scvelo.preprocessing.utils import not_yet_normalized, normalize_per_cell
from scvelo.preprocessing.neighbors import neighbors, get_connectivities, get_n_neighs, verify_neighbors
from scipy.sparse import csr_matrix, issparse
```

```python
def knn_smoothing(adata_rna, layer="raw", use_rep="X_pca", n_neighbors=30, n_pcs=40, method="umap",mode='connectivities'):
    '''
    Returns anndata object with smoothed version of adata_rna.layers[layer] as a new layer
    '''
    ## Adapting from scvelo code
    neighbors(adata_rna, n_neighbors=n_neighbors, use_rep=use_rep, n_pcs=n_pcs, method=method)
    connectivities = get_connectivities(
                adata_rna, mode, n_neighbors=n_neighbors, recurse_neighbors=False
            )
    adata_rna.layers["KNNsmooth_" + layer] = (
        csr_matrix.dot(connectivities, csr_matrix(adata_rna.layers[layer]))
        .astype(np.float32)
        .A
    )
    return(adata_rna)
```

```python
normalize_per_cell(adata_rna)
adata_rna = knn_smoothing(adata_rna, n_neighbors=30)
```
